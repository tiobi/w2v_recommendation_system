```python
import pandas as pd
import glob
import numpy as np
import nltk
import csv
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import RegexpTokenizer, word_tokenize, sent_tokenize
from gensim.models.word2vec import Word2Vec
from konlpy.tag import Twitter
import re

file_name = "total_robot_team_d2.csv"
f = open(file_name, 'r', encoding = 'utf-8')
data = csv.reader(f)
f.close
next(data)
sns_text = []
for row in data:
    sns_text.append(str(row[-1]))
    
tokenized_text = []
for lines in sns_text:
    tokenized_text.append(sent_tokenize(str(sns_text)))
embedding_model = Word2Vec(tokenized_text, size = 100, window = 2, min_count = 10, workers = 2, iter= 100, sg = 1) 
```

```python
embedding_model = Word2Vec(tokenized_text, size = 100, window = 2, min_count = 5, workers = 2, iter= 100, sg = 1) 
```
